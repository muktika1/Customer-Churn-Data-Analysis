```python
def CountPlot_Table (feature):
    # Create Count Plot for Churn Vs Feature
    sns.countplot(x=feature, hue="Churn", data=df, palette="Paired", edgecolor = 'Black', order=df[feature].value_counts().index)
    sns.despine()
    # Create a plot for proportions
    temp_table = pd.DataFrame(round(df.groupby(feature)['Churn'].value_counts(normalize = True),4))
    table = plt.table(cellText=temp_table.values,
          rowLabels=temp_table.index,
          colLabels=temp_table.columns,
          bbox=(1.5, 0,0.4 , 0.45))
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    plt.show()
```

```python
#DEMOGRAPHIC DEPENDENCE ON CHURN
demo_features = ['gender','SeniorCitizen','Partner','Dependents']
for feature in demo_features:
    CountPlot_Table(feature)
```

```python
# CHURN BASED ON SERVICES AVAILED
demo_features = ['PhoneService','MultipleLines','InternetService','StreamingTV','StreamingMovies',
     'OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport']
for feature in demo_features:
    CountPlot_Table(feature)
```

```python
# CHURN RATE BASED ON BILLING METHOD
demo_features = ['Contract','PaperlessBilling','PaymentMethod']
for feature in demo_features:
    CountPlot_Table(feature)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, classification_report, confusion_matrix, f1_score, precision_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
```

# Baseline models

### Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
log_y_pred = log_reg.predict(X_test)

log_recall      = recall_score(y_test, log_y_pred)
log_accuracy    = accuracy_score(y_test, log_y_pred)
log_conf_martix = confusion_matrix(y_test, log_y_pred)
log_f1          = f1_score(y_test, log_y_pred)
log_precision   = precision_score(y_test, log_y_pred, average='macro')

print ("Recall score    ", log_recall)
print ("Accuracy score  ", log_accuracy)
print ("F1 Score        ", log_f1)
print ("Precision Score ", log_precision)
print ("Confusion Matrix\n", log_conf_martix) #not sure if we need this matrix
```

### Gaussian Naive Bayes

```python
gaussian_clf = GaussianNB()
gaussian_clf.fit(X_train, y_train)
gaussian_y_pred = gaussian_clf.predict(X_test)

guassian_recall      = recall_score(y_test, gaussian_y_pred)
gaussian_accuracy    = accuracy_score(y_test, gaussian_y_pred)
gaussian_conf_matrix = confusion_matrix(y_test, gaussian_y_pred)
gaussian_f1          = f1_score(y_test, gaussian_y_pred)
gaussian_precision   = precision_score(y_test, gaussian_y_pred, average='macro')

print ("Recall score    ", guassian_recall)
print ("Accuracy score  ", gaussian_accuracy)
print ("F1 Score        ", gaussian_f1)
print ("Precision Score ", gaussian_precision)
print ("Confusion Matrix\n", gaussian_conf_matrix)
```

### Random Forest Classifier

```python
random_forest_clf = RandomForestClassifier()
random_forest_clf.fit(X_train, y_train)
random_forest_y_pred = random_forest_clf.predict(X_test)

random_forest_recall      = recall_score(y_test, random_forest_y_pred)
random_forest_accuracy    = accuracy_score(y_test, random_forest_y_pred)
random_forest_conf_matrix = confusion_matrix(y_test, random_forest_y_pred)
random_forest_f1          = f1_score(y_test, random_forest_y_pred)
random_forest_precision   = precision_score(y_test, random_forest_y_pred)

print ("Recall score    ", random_forest_recall)
print ("Accuracy score  ", random_forest_accuracy)
print ("F1 Score        ", random_forest_f1)
print ("Precision Score ", random_forest_precision)
print ("Confusion Matrix\n", random_forest_conf_matrix)

```

### XGBoost

```python
xgb_clf = XGBClassifier()
xgb_clf.fit(X_train, y_train)
xgb_y_pred = xgb_clf.predict(X_test)

xgb_recall      = recall_score(y_test, xgb_y_pred)
xgb_accuracy    = accuracy_score(y_test, xgb_y_pred)
xgb_conf_matrix = confusion_matrix(y_test, xgb_y_pred)
xgb_f1          = f1_score(y_test, xgb_y_pred)
xgb_precision   = precision_score(y_test, xgb_y_pred)

print ("Recall score    ", xgb_recall)
print ("Accuracy score  ", xgb_accuracy)
print ("F1 Score        ", xgb_f1)
print ("Precision Score ", xgb_precision)
print ("Confusion Matrix\n", xgb_conf_matrix)
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
# Define the search space
param_grid = {
    # Percentage of columns to be randomly samples for each tree.
    "colsample_bytree": [ 0.3, 0.5 , 0.8 ],
    # reg_alpha provides l1 regularization to the weight, higher values result in more conservative models
    "reg_alpha": [0, 0.5, 1, 5],
    # reg_lambda provides l2 regularization to the weight, higher values result in more conservative models
    "reg_lambda": [0, 0.5, 1, 5]
    }
# Set up score
scoring = ['recall']
# Set up the k-fold cross-validation
kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)
```

```python
# Define grid search
grid_search = GridSearchCV(estimator=xgb_clf,
                           param_grid=param_grid,
                           scoring=scoring,
                           refit='recall',
                           n_jobs=-1,
                           cv=kfold,
                           verbose=3)
# Fit grid search
grid_result = grid_search.fit(X_train_smote, y_train_smote)
# Print grid search summary
grid_result
# Print the best score and the corresponding hyperparameters
print(f'The best score is {grid_result.best_score_:.4f}')
print('The best score standard deviation is', round(grid_result.cv_results_['std_test_recall'][grid_result.best_index_], 4))
print(f'The best hyperparameters are {grid_result.best_params_}')
```
